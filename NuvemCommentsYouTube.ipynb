```python
# importar dados e organizar
import pandas as pd
import numpy as np

# plotagem de imagens e ajuste de cores
import matplotlib.pyplot as plt
import matplotlib as mpl
from wordcloud import  WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```python
# lendo o arquivo
data = pd.read_csv('DataCommentsYoutube.csv', encoding='utf-8')
```

```python
data.head() # saida do arquivo
```

```python
text = data.Comentários.str.lower().tolist()  # cria uma lista apenas com os comentários
```

```python
STOPWORDS = ['um','no', 'box', 'revestimento', 'obrigada', 'pequeno', 'tem', 'do', 'oi', 'apenas', 'pastilha','porcelanato', 'nichos','gostaria', 'saber', 'olá', 'quot', 'no', 'pra', 'fazer','video', 'dica', 'banheiro','espelho','banheiros', 'parede', 'tenho', 'minha', 'porta', 'menino' 'espelho', 'ideia', 'nicho', 'dicas', 'vc', 'todo', 'contato', 'sou', 'meu', 'em', 'para', 'bem', 'uma', 'e', 'se', 'os', 'fiz', 'que', 'da', 'com', 'qual', 'pode', 'eu', 'qual', 'por', 'ver','principal','essa','vez','nas','mas','qual','principal','ele','ter','doença','pois','este','vez','ver principal','artigo principal','já','aos','pode','outro','artigo','desse','alguns','meio','entre','das','podem','esse','seu','também','são','quando','de', 'que','em','os','as','da','como','dos','ou','se','um','uma','para','na','ao','mais','por','não','ainda','muito','sua', 'vídeo', 'estou','me','fica', 'só', 'colocar', 'boa', 'noite', 'dia', 'posso']
# palavara para ignorar 
```

```python
words = ' '.join(text) # cria uma blocão de palavras
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS).generate(words)
```

```python
plt.imshow(wordcloud, interpolation='bilInear')  # visualização das palavras
plt.axis('off')
plt.figure( figsize=(80,40))
plt.show()
```
